```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("AmazonBooks - Sheet1.csv")
df.head(4)
```

```python
df.tail(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(15, 5))
sns.set_theme(style="darkgrid")
sns.countplot(x='Genre', data=df)
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.histplot(x=df['User Rating'], kde=True, fill=True)
plt.title("User Rating distribution")
plt.show()
```

```python
g = sns.displot(x=df['User Rating'], hue=df['Genre'], kind="kde", fill=True)
plt.title("User Rating distribution by Genre")
g.fig.set_figerwidth(15)
g.fig.set_figheight(5)
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.boxenplot(x="Year", y="User Rating", data=df)
plt.title('User rating distribution by year')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.histplot(x="Price", data=df, fill="True", kde=True)
plt.title("Price distribution")
plt.show()
```

```python
g = sns.displot(x=df['Price'], hue=df['Genre'], kind="kde", fill=True)
plt.title("Price distribution by Genre")
g.figure.set_figwidth(15)
g.figure.set_figheight(5)
plt.show()
```

```python
df.query('Price > 75')
```

```python
plt.figure(figsize=(15, 5))
sns.pointplot(x="Year", y="Price", hue="Genre", data=df)
plt.title('Price distribution by year and genre')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.histplot(x="Reviews", data=df, fill="True", kde=True)
plt.title("Reviews distribution")
plt.show()
```

```python
df.query("Reviews > 90000")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

sns.scatterplot(x="Price", y="Reviews", hue='Genre', data=df, ax=ax[0])
sns.scatterplot(x="Price", y="User Rating", hue='Genre', data=df, ax=ax[1])
plt.show()
```
